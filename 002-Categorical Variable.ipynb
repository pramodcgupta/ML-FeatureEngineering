# Handling Categorical Variables

```python
#------------------------------------------------------------------------------------
# Pramodkumar Gupta
# Handing Numerical Features
#------------------------------------------------------------------------------------

# Import all libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
use_cols = ['id', 'purpose', 'loan_status', 'home_ownership']
df=pd.read_csv('loan.csv', usecols=use_cols).sample(10000, random_state=44)
df.head()
```

```python
# Lets Analyse Categorical variable
```

```python
# Bar Plot
df['home_ownership'].value_counts().plot.bar()
plt.title('Home Ownership')
plt.xlabel('Home Ownership')
plt.ylabel('Counts')
```

```python
df['home_ownership'].value_counts()
```

```python
# Bar Plot
df['purpose'].value_counts().plot.bar()
plt.title('purpose')
plt.xlabel('purpose')
plt.ylabel('Counts')
```

```python
# Bar Plot
df['loan_status'].value_counts().plot.bar()
plt.title('loan_status')
plt.xlabel('loan_status')
plt.ylabel('Counts')
```

```python
# You can also plot countplot to analyse Categorical variable
import seaborn as sns
sns.countplot(x = 'home_ownership', data = df)
```
